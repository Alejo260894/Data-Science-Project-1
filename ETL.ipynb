Realizaremos la limpieza de datos de los tres Data Frames generados

```python
#Vizualiso la informacion del data frame
df_games.info()
```

```python
#Eliminar las filas que tienen solo nulo en todas sus columnas
steam_games = df_games.dropna(how='all')
steam_games.info()
```

```python
#Cuento la cantidad de Nulos
steam_games.isnull().sum()
```

Valido que las tres filas no cuentan con informacion relevante por lo que procedo a borrarlas, adicionalmente convierto el tipo valor de id en enteros

```python
steam_games.dropna(subset=['id'],inplace=True)
steam_games.dropna(subset=['app_name'], inplace=True)
steam_games['id'] = steam_games['id'].astype('int')
steam_games.info()
```

Revisando los datos me doy cuenta que la columna Publisher y Developer son parecidos por lo que tomo los valores de Publisher para llenar la informacion faltante en Developer y viceversa


```python
steam_games['publisher'].fillna(steam_games['developer'],inplace=True)
steam_games['developer'].fillna(steam_games['publisher'],inplace=True)
steam_games.isnull().sum()
```

Se procede a transformar y limpiar las columnas segun los requerimientos de la API

```python
# Se observan las cantidades de registros por cada fecha
steam_games['release_date'].value_counts()
```

```python
# Convertir la columna 'release_date' al tipo datetime
steam_games['release_date'] = pd.to_datetime(steam_games['release_date'], errors='coerce')
# Extraer el año y crear una nueva columna 'Año'
steam_games['año_lanzamiento'] = steam_games['release_date'].dt.year   
steam_games['año_lanzamiento'] = steam_games['año_lanzamiento'].astype('Int64')
# elimina la columna 'release_date'
steam_games = steam_games.drop('release_date', axis=1)

# Mostrar el dataframe resultante
steam_games.head()
```

```python
# Suponiendo que df es tu DataFrame y 'precio' es la columna de precios
# Reemplazar valores de texto por 0
steam_games['price'] = pd.to_numeric(steam_games['price'], errors='coerce')
steam_games['price'] = steam_games['price'].replace(['Free to Use', 'Free to Play'], 0)

# Imputar valores nulos como 0
steam_games['price'] = steam_games['price'].fillna(0)

# Convertir a tipo numérico (si no lo ha hecho automáticamente)
steam_games['price'] = pd.to_numeric(steam_games['price'])
```

Se tiene que separar los generos ubicados en listas por lo que se copiaran las filas segun los generos que haya en cada lista y luego se borraran las filas que tengan generos vacios

```python
steam_games = steam_games.explode('genres')
steam_games = steam_games.dropna(subset=['genres'])
steam_games.head(20)
```

```python
steam_games.info()
```

```python
#Eliminar columnas que no se utilizaran
steam_games = steam_games.drop(['tags', 'specs', 'url', 'reviews_url'], axis=1)
steam_games.columns
```

```python
#le asigno el string 'Sin Datos' a las columnas publiser y developer que no cuentan con informacion
steam_games['publisher'].fillna(value='Sin datos', inplace=True)
steam_games['developer'].fillna(value='Sin datos', inplace=True)
steam_games.info()
```

```python
#Subo el archivo corregido
archivo_limpio = 'Datasets_limpio/steam_games.csv'
steam_games.to_csv(archivo_limpio, index=False, encoding='utf-8')
```
